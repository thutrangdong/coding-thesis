```python
from pytrends.request import TrendReq
import datetime
import pandas as pd
import matplotlib as plt
import requests
import json
%matplotlib inline

response = requests.get('https://api.coindesk.com/v1/bpi/historical/close.json?start=2010-07-17&end=2017-12-05')
json_data = json.loads(response.content)
bpi = pd.DataFrame.from_dict(json_data)
bpi.drop(['disclaimer', 'time'], axis=1, inplace=True)
bpi.drop(['updated', 'updatedISO'], inplace=True)
bpi.index.name = 'date'
bpi.plot()
```
